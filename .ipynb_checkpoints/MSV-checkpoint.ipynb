We consider a multivariate stochastic volatility (MSV) model. 
Let $\Sigma$ be a $p\times p$ matrix with $\Sigma_{ij} = \rho^{|i-j|}$, $X^{(1)} \sim \mathcal N(0,\Sigma)$ and
\begin{equation*}
\begin{aligned}
X^{(t)}\mid X^{(1:t-1)}&\sim\mathcal N\left(\alpha X^{(t-1)},\Sigma\right),\\
Y^{(t)}\mid X^{(1:t)},Y^{(1:t-1)}&\sim\mathcal N\left(0,\beta^2 diag(\exp(X^{(t)}))\right),
\end{aligned}
\end{equation*}
where $\exp(X^{(t)})$ represents the elementwise application of the exponential function to the vector $X^{(t)}$. If $p=1$, this reduces to the standard SV model. In this section, the MSE is taken with respect to the oracle posterior mean $\mathbb E[X^{(t)}\mid Y^{(1:t)}]$ (obtained by running SMC with $3,000$ particles and $4$-descendant SMG) instead of the true value $X^{(t)}$, since otherwise the difference between the posterior mean and the true value would dominate the error.

```python
# Set parameters
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
alpha = .7
rho = .8
dim = 2
nParticle = 30
nT  = 50
beta = 1
Sigma = np.zeros([dim, dim])
for i in range(dim):
    for j in range(dim):
        Sigma[i,j] = rho**np.abs(i-j)
```
